```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout, LSTM
```

```python
def load_dataset(folder_path, num_classes=10):
    texts = []
    labels = []
    for label in range(num_classes):
        file_path = os.path.join(folder_path, f"{label}.txt")
        with open(file_path, "r", encoding="latin-1") as f:
            lines = [line.strip() for line in f if line.strip()]
        texts.extend(lines)
        labels.extend([label] * len(lines))
    return texts, np.array(labels)

train_texts, y_train = load_dataset("data/tobacco-data")
test_texts, y_test = load_dataset("data/sfsf_test_pol_data_bbf")

print("Train samples:", len(train_texts))
print("Test samples:", len(test_texts))
print("Example text:", train_texts[0][:200])
```

```python
max_words = 20000   # vocabulary size
max_len = 200       # max number of tokens per text

tokenizer = Tokenizer(num_words=max_words, oov_token="<UNK>")
tokenizer.fit_on_texts(train_texts)

X_train = pad_sequences(tokenizer.texts_to_sequences(train_texts), maxlen=max_len)
X_test  = pad_sequences(tokenizer.texts_to_sequences(test_texts),  maxlen=max_len)

vocab_size = min(max_words, len(tokenizer.word_index) + 1)

print("Vocab size:", vocab_size)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
embedding_dim = 100

cnn_model = Sequential([
    Embedding(input_dim=vocab_size,
              output_dim=embedding_dim,
              input_length=max_len),
    Conv1D(128, 5, activation="relu"),
    GlobalMaxPooling1D(),
    Dense(64, activation="relu"),
    Dropout(0.5),
    Dense(10, activation="softmax")  # 10 classes: 0–9
])

cnn_model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

cnn_model.summary()
```

```python
history_cnn = cnn_model.fit(
    X_train, y_train,
    batch_size=64,
    epochs=5,          # you can increase later if you want
    validation_split=0.2,
    verbose=1
)
```

```python
cnn_loss, cnn_acc = cnn_model.evaluate(X_test, y_test, verbose=0)
print("CNN Test accuracy:", cnn_acc)
```

```python
lstm_model = Sequential([
    Embedding(input_dim=vocab_size,
              output_dim=embedding_dim,
              input_length=max_len),
    LSTM(128, dropout=0.2, recurrent_dropout=0.2),
    Dense(64, activation="relu"),
    Dropout(0.5),
    Dense(10, activation="softmax")
])

lstm_model.compile(
    loss="sparse_categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

lstm_model.summary()
```

```python
history_lstm = lstm_model.fit(
    X_train, y_train,
    batch_size=64,
    epochs=5,
    validation_split=0.2,
    verbose=1
)

lstm_loss, lstm_acc = lstm_model.evaluate(X_test, y_test, verbose=0)
print("LSTM Test accuracy:", lstm_acc)
```

```python
print("CNN Test accuracy: ", cnn_acc)
print("LSTM Test accuracy:", lstm_acc)
```
